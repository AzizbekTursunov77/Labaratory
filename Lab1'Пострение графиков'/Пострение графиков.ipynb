```python
from numpy import *
import matplotlib.pyplot as plt
```

```python
for x in [1, 10, 1000]:
    y = log( exp(1 / (sin(x) + 1)) / (5/4 + 1/x**15) ) / log(1 + x*x)
    print("x = " + str(x) + ", y = " + str(y))
```

Упражнение 2

```python
x = arange(-5, 6.01, 0.01)

plt.figure(figsize = (10, 10))
plt.plot(x, x*x - x - 6)
plt.axhline(y = 0, color = 'r', linestyle = '-')
plt.title(r'$x_1 = -2; \ x_2 = 3$', fontsize = 15) 
plt.grid(True)
```

Упражнение 3

```python
x = arange(-10, 10.01, 0.01)
y = log(x**2 + 1) / log(1 + tan(1 / (1 + sin(x)**2))) * exp(-abs(x)/10)

plt.figure(figsize = (10, 10))
plt.plot(x, y)
plt.grid(True)
plt.show()
```

Упражнение 4

```python
#Используя функцию eval() постройте график функции, введённой с клавиатуры
function = input("Введите функцию от x: ")
xmin = input("Введите минимальный x: ")
xmax = input("Введите максимальный x: ")
step = input("Введите шаг счета: ")
x = arange(float(xmin), float(xmax), float(step))
y = eval(function)

plt.figure(figsize = (10, 10))
plt.plot(x, y)
plt.grid(True)
plt.show()


#эффект «рисование от руки»
with plt.xkcd():
    plt.figure(figsize = (10, 10))
    plt.plot(x, y)
```

```python

a = 0
while a % 2 == 0 or a == 1 or a < 0:
    a = int(input("Введите нечетное a, не равное 1: a = "))
b = 0
while b <= 0 or b >= 1:
    b = float(input("Введите положительное число, меньшее 1: b = "))

xmin = float(input("Введите минимальный x: "))
xmax = float(input("Введите максимальный x: "))
step = float(input("Введите шаг счета: "))
x = arange(xmin, xmax, step)
n0 = int(100 / (xmax - xmin))
print("=================== " + str(n0) + " ===================")

j = 0
y = []
for x0 in x:
    y.append(0)

    for n in range(n0 + 1):
        y[j] += b**n * cos(a**n * pi * x0)
    
    j += 1

plt.figure(figsize = (15, 10))
plt.plot(x, y)

plt.show()
```
